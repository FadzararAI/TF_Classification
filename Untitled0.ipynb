```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Activation,Flatten,Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Image preprocessing using Image Segmentation
train_data = ImageDataGenerator(rescale=1./255,validation_split=0.4,rotation_range=20,horizontal_flip=True,shear_range=0.2,fill_mode='wrap')
```

```python
# Creates training data
train_gen = train_data.flow_from_directory('/content/drive/My Drive/Colab Notebooks/rockpaperscissors/rps-cv-images', target_size=(100,150),class_mode='categorical',subset='training')
# Creates validation data
valid_gen = train_data.flow_from_directory('/content/drive/My Drive/Colab Notebooks/rockpaperscissors/rps-cv-images', target_size=(100,150),class_mode='categorical',subset='validation')
```

```python
# Training dataset overview
for _ in range(3):
  img, label = train_gen.next()
  plt.imshow(img[0])
  plt.show()
```

```python
# Validation dataset overview
for _ in range(3):
  img, label = valid_gen.next()
  plt.imshow(img[0])
  plt.show()
```

```python
# Creating 3 layers as well as the optimizations
def layers(w,h,d,c):
  model = tf.keras.models.Sequential()
  inshape = (w,h,d)
  # First layer
  model.add(Conv2D(32,(3,3),input_shape=inshape))
  model.add(Activation("relu"))
  model.add(MaxPooling2D(2,2))
  # Second layer
  model.add(Conv2D(64,(3,3)))
  model.add(Activation("relu"))
  model.add(MaxPooling2D(2,2))
  # Third layer
  model.add(Conv2D(128,(3,3)))
  model.add(Activation("relu"))
  model.add(MaxPooling2D(2,2))
  # Activation function using softmax
  model.add(Flatten())
  model.add(Dense(c))
  model.add(Activation("softmax"))
  # Optimize
  model.compile(optimizer=Adam(), loss="categorical_crossentropy", metrics=['accuracy'])
  return model
  
```

```python
model_training = layers(100,150,3,3).fit(train_gen,validation_data=valid_gen,verbose=2,epochs=30,steps_per_epoch=25,validation_steps=5)
```

```python
# Shows the accuracy
model_training.history['accuracy'][-1] * 100
```
